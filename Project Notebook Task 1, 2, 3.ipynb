# Project 2: EDA and Feature Engineering

**Time Based Trends**

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["Month"] = df["Date"].dt.to_period("M")
```

```python
monthly_sales = df.groupby("Month")["TotalPrice"].sum()
print(monthly_sales)
```

```python
monthly_sales.plot(kind="line", marker="o", figsize=(10, 5), title="Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

**Outlier Detection (IQR)**

```python
def iqr_outliers(series):
  q1 = series.quantile(0.25)
  q3 = series.quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5 * iqr
  upper_bound = q3 + 1.5 * iqr
  return series[(series < lower_bound) | (series > upper_bound)]

for col in ["Quantity", "TotalPrice", "UnitPrice", "ItemsInCart"]:
  outliers = iqr_outliers(df[col])
  print(f"{col}: {len(outliers)} outliers detected out of {len(df)} rows")

numeric_cols = ["Quantity", "TotalPrice", "UnitPrice", "ItemsInCart"]
fig, axes = plt.subplots(1, len(numeric_cols), figsize=(4 * len(numeric_cols), 4))
for ax, col in zip(axes, numeric_cols):
    df.boxplot(column=col, ax=ax)
    ax.set_title(col)
plt.tight_layout()
plt.show()
```

```python
from scipy import stats

z_scores = np.abs(stats.zscore(df["TotalPrice"]))
zscore_outliers = df[z_scores > 3]
print(f"Outliers detected: {len(outliers)}")
```

**Correlation Analysis**

```python
corr_matrix = df[numeric_cols].corr()
print(corr_matrix)

plt.figure(figsize=(6, 5))
plt.imshow(corr_matrix, cmap="coolwarm", vmin=-1, vmax=1)
plt.colorbar()
plt.xticks(range(len(numeric_cols)), numeric_cols, rotation=45, ha="right")
plt.yticks(range(len(numeric_cols)), numeric_cols)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

**Large Order Indicator**

```python
df["LargeOrder"] = (df["Quantity"] >= 5).astype(int)
```

**Customer Order Frequency**

```python
df["CustomerOrderCount"] = df.groupby("CustomerID")["OrderID"].transform("count")
```

**Customer Lifetime Spending**

```python
df["CustomerLifetimeValue"] = df.groupby("CustomerID")["TotalPrice"].transform("sum")
```

**Average Customer Order Value**

```python
df["AvgCustomerOrderValue"] = (
    df.groupby("CustomerID")["TotalPrice"]
      .transform("mean")
)
```

**Number of Products Purchased by Customer**

```python
df["CustomerTotalItems"] = (
    df.groupby("CustomerID")["Quantity"]
      .transform("sum")
)
```

```python
df["LargeOrder"].value_counts()
```

```python
df["IsWeekend"] = df["Date"].dt.dayofweek.isin([5,6]).astype(int)
```

```python
import matplotlib.pyplot as plt

df.groupby("IsWeekend")["TotalPrice"].mean().plot(kind="bar")
plt.xlabel("Weekend")
plt.ylabel("Average Total Price")
plt.show()
```

# Project 3: SQL Analysis

```python
import sqlite3

# Create an in-memory (or file-based) SQLite database
conn = sqlite3.connect("orders.db")

# Convert the 'Month' column to string as SQLite does not support Period type
df_sql = df.copy()
df_sql['Month'] = df_sql['Month'].astype(str)

# Push your cleaned dataframe into a SQL table
df_sql.to_sql("orders", conn, if_exists="replace", index=False)
```

```python
pd.read_sql_query("SELECT * FROM orders LIMIT 5;", conn)
```

```python
pd.read_sql_query("""SELECT OrderID, Product, TotalPrice
FROM orders
WHERE TotalPrice > 1500;""", conn)
```

```python
pd.read_sql_query("""SELECT *
FROM orders
WHERE OrderStatus = 'Cancelled';""", conn)
```

```python
pd.read_sql_query("""SELECT OrderID, Product, TotalPrice
FROM orders
ORDER BY TotalPrice DESC
LIMIT 10;""", conn)
```

```python
#Revenue and order count per product
pd.read_sql_query("""SELECT Product,
       COUNT(*) AS num_orders,
       SUM(TotalPrice) AS total_revenue,
       AVG(TotalPrice) AS avg_order_value
FROM orders
GROUP BY Product
ORDER BY total_revenue DESC;""", conn)
```

```python
#Order status breakdown
pd.read_sql_query("""SELECT OrderStatus,
       COUNT(*) AS count
FROM orders
GROUP BY OrderStatus
ORDER BY count DESC;""", conn)
```

```python
#Most common referral source
pd.read_sql_query("""SELECT ReferralSource,
       COUNT(*) AS occurrences
FROM orders
GROUP BY ReferralSource
ORDER BY occurrences DESC;""", conn)
```

```python
#Average order value by payment method, only for delivered orders
pd.read_sql_query("""SELECT PaymentMethod,
       AVG(TotalPrice) AS avg_value, COUNT(*) AS num_orders
FROM orders
WHERE OrderStatus = 'Delivered'
GROUP BY PaymentMethod
ORDER BY avg_value DESC;""", conn)
```

```python
pd.read_sql_query("""SELECT strftime('%Y-%m', Date) AS month,
       SUM(TotalPrice) AS monthly_revenue
FROM orders
GROUP BY month
ORDER BY monthly_revenue DESC
LIMIT 5;""", conn)
```
